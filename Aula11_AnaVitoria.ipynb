```

#Def de Jacobi, apenas com as interações
#Nesse exemplo do livro, o limite da 17 mas, nesse caso, não temos um x estipulado e nem um erro


import numpy as np
from numpy import diag
from numpy import diagflat

def Jacobi_iteration (a, b, d, x):
  
  n = len(a)
  lim = 0
  
  while((lim < 17)): 
    for i in range(n):
      s = 0
      for j in range(n):
        if(j != i): #Aqui ele entra no looping apenas se j for diferente de i, ou seja, ele não vai considerar as diagonais
          s = s + (a[i][j] * x[j]) #realiza a soma de todas as multiplicações dos valores da linha i com o de j de x(a matriz "base")
      x[i] = (b[i]-s)/d[i] #O d é a diagonal armazenada antes de chamar a função
    print("interação",lim,":", x)  
    lim = lim + 1
    
  return a,x



a = np.array([[10.,-1.,2.,0.],[-1.,11.,-1.,3.],[2.,-1.,10.,-1.],[0.,3.,-1.,8.]])
b = np.array([6.,25.,-11.,15.])
x = n*[1]

n = len(a)

# Achei melhor separar a diagonal da matriz e vice-versa
# diag e diagflat para coletar e retirar a diagonal
diagonal_de_a = diag(a) 
matriz_sem_diag = a - diagflat(diagonal_de_a)

print("Matriz sem a diagonal:")
print("\n",matriz_sem_diag,"\n")


a,x = Jacobi_iteration(matriz_sem_diag, b, diagonal_de_a, x)

```
